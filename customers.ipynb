```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv("customers.csv", encoding="latin-1")
```

```python
customers.isnull().sum()
```

```python
customers.drop(range(2241,4999),inplace=True)
```

```python
customers.isnull().sum()
```

```python
Customers=customers.drop(["Ever_Married","Family_Size",],axis=1,inplace=True)
customers.columns
```

```python
customers.isnull().sum()
```

```python
customers.info()
```

```python
sns.histplot(customers['Age'], bins=30, kde=True)
plt.title("Customer Age Distribution")
plt.show()
```

```python
sns.barplot(x='Segmentation', y='Spending_Score', data=customers)
plt.title("Spending by Segment")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Spending_Score', hue='Segmentation', data=customers)
plt.title("Age vs Spending")
plt.show()
```

```python
sns.countplot(x='Segmentation', data=customers)
plt.title("Customer Segments Count")
plt.show()
```

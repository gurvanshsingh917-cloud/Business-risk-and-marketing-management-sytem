```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn = pd.read_csv("churn.csv", encoding="latin-1")
```

```python
churn.isnull().sum()
```

```python
churn.drop(range(2241,9994),inplace=True)
```

```python
churn.isnull().sum()
```

```python
churn["Total Spending"].fillna(churn["Total Spending"].mean(0),inplace=True)
```

```python
churn.isnull().sum()
```

```python
sns.countplot(x='Churn', data=churn)
plt.title("Churn Count")
plt.show()
```

```python
sns.boxplot(x='Churn', y='Monthly Spending', data=churn)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.countplot(x='Gender', hue='Churn', data=churn)
plt.title("Contract Type vs Churn")
plt.show()
```
